# INNER JOIN

```python
query='select * from customer c inner join order_table o on c.cust_id=o.cust_id'
item=pd.read_sql(query,engine)
item
```

```python
query='''select c.cust_name, c.cust_id, count(c.cust_id) 
from customer c inner join order_table o on c.cust_id=o.cust_id
group by c.cust_id order by count(c.cust_id) desc;'''
ot1=pd.read_sql(query,engine)
ot1
```

```python
query='select * from customer c left join order_table o on c.cust_id=o.cust_id'
join1=pd.read_sql(query,engine)
join1
```

```python
query='select * from customer c right join order_table o on c.cust_id=o.cust_id'
join2=pd.read_sql(query,engine)
join2
```

```python
query='select * from customer c full outer join order_table o on c.cust_id=o.cust_id'
join3=pd.read_sql(query,engine)
join3
```

```python
query='select * from customer c inner join order_table o on c.cust_id=o.cust_id and o.quanti>2'
item=pd.read_sql(query,engine)
item
```

```python
query='select * from customer c inner join order_table o on c.cust_id=o.cust_id and o.item_id in (3,9,12,16,20)'
item=pd.read_sql(query,engine)
item
```
